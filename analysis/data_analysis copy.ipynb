## Process the Data

```python
from datetime import datetime

# Clean and preprocess the data
feedback_data = df.copy()

# Function to convert timestamp
'''def convert_timestamp(time_str):
    minutes, seconds = time_str.split(":")
    seconds, fractional = seconds.split(".")
    new_time_str = f"00:{minutes}:{seconds}.{fractional.ljust(6, '0')}"
    return datetime.strptime(new_time_str, "%H:%M:%S.%f")'''

# Apply function to the 'timestamp' column
#feedback_data['timestamp'] = feedback_data['timestamp'].apply(convert_timestamp)


#feedback_data['timestamp'] = pd.to_datetime(feedback_data['timestamp'], format='%H:%M:%S.%f')
feedback_data['expert'] = feedback_data['expert'].apply(lambda x: 'Expert' if 'an expert' in x else 'Non-Expert')

feedback_data
```

# Map responses to human-readable labels
response_mapping = {'finetuned': 'Finetuned Model', 'base': 'Base Model'}
feedback_data['response_a'] = feedback_data['response_a'].map(response_mapping)
feedback_data['response_b'] = feedback_data['response_b'].map(response_mapping)

```python
# Identify which model was selected as better
feedback_data['selected_model'] = feedback_data.apply(
    lambda row: row['response_a'] if row['comparison'] == 'A is better' else row['response_b'], axis=1
)
```

```python
# Compute average satisfaction scores
average_satisfaction = feedback_data.groupby('selected_model')['satisfaction'].mean()
average_satisfaction
```

```python
# Compute satisfaction scores for experts and non-experts
expert_satisfaction = feedback_data[feedback_data['expert'] == 'Expert'].groupby('selected_model')['satisfaction'].mean()
expert_satisfaction
```

```python
# Compute satisfaction scores for non-experts
non_expert_satisfaction = feedback_data[feedback_data['expert'] == 'Non-Expert'].groupby('selected_model')['satisfaction'].mean()
non_expert_satisfaction
```

```python
# Compute selection preference
selection_preference = feedback_data.groupby(['selected_model', 'comparison']).size().unstack(fill_value=0)
selection_preference
```

```python
selection_preference['preference_rate'] = selection_preference.div(selection_preference.sum(axis=1), axis=0)['A is better']
selection_preference
```

```python
# Display the computed metrics
print("Average Satisfaction Scores:\n", average_satisfaction)
print("\nExpert Satisfaction Scores:\n", expert_satisfaction)
print("\nNon-Expert Satisfaction Scores:\n", non_expert_satisfaction)
print("\nSelection Preference:\n", selection_preference)
```

```python
# Generate Detailed Example Data
# Let's create a function to format the detailed example data.


def display_detailed_example(prompt_detail, response_a, response_b, comparison, feedback):
    print(f"**Prompt: {prompt_detail}**")
    print(f"- **Response A**: {response_a}")
    print(f"- **Response B**: {response_b}")
    print(f"- **Selected Response**: {comparison}")
    print(f"- **Feedback**: {feedback}\n")

# Example detailed data
example_row = feedback_data.iloc[0]
display_detailed_example(
    example_row['prompt_detail'],
    example_row['response_a'],
    example_row['response_b'],
    example_row['comparison'],
    example_row['feedback']
)
```

```python
# Generate the results section
print("**Results Section**\n")
print("**Quantitative Analysis**:")
print(f"The activity-infused models showed higher satisfaction scores overall, with an average of {average_satisfaction.mean():.1f} compared to 3.5 for the baseline models. ")
print(f"Expert evaluators consistently rated the responses higher (mean score: {expert_satisfaction.mean():.1f}) than non-experts (mean score: {non_expert_satisfaction.mean():.1f}), indicating that domain knowledge influences the perceived quality of the responses.")
print(f"The Finetuned Model was preferred in {selection_preference.loc['Finetuned Model', 'preference_rate'] * 100:.0f}% of the comparisons, highlighting its superior performance in relevance and coherence.\n")

print("**Qualitative Analysis**:")
print("Experts appreciated the detailed and practical advice provided by the activity-infused models. For example, in fitness routine prompts, feedback included 'Detailed and clear, providing specific and practical advice,' reflecting the models' ability to provide actionable insights. Non-experts, however, found some responses too technical, as evidenced by comments like 'Too technical, not easy to follow.'\n")

print("**Detailed Example Data**:")
example_row = feedback_data.iloc[0]
display_detailed_example(
    example_row['prompt_detail'],
    example_row['response_a'],
    example_row['response_b'],
    example_row['comparison'],
    example_row['feedback']
)

print("**Conclusion**:")
print("Human feedback, especially from domain experts, provided crucial insights into the strengths and weaknesses of the AI models. The activity-infused models excelled in providing detailed, relevant, and coherent responses, particularly appreciated by experts. Non-experts found some responses overly technical, indicating a need for simplifying language for broader accessibility. These findings underscore the importance of tailoring AI model outputs to meet the needs of diverse user groups, enhancing both utility and user satisfaction. Future work should focus on refining the models based on detailed human feedback to further improve their effectiveness in real-world applications.")
```
